```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
import pandas as pd
from PIL import Image
from PIL.ExifTags import TAGS

path="/content/drive/My Drive/Final Year Research/Dataset/Oneplus 5T/Level 04/*"
len_str_remove=len(path)-1
df = pd.DataFrame(columns=['image', 'brightness','shutter_speed','iso'])
i=0
for file in glob.glob(path):
    #print(file[len_str_remove:])
    img = Image.open(file)
    # extract EXIF data
    exifdata = img.getexif()
    img_mdata=np.zeros(3)
    # iterating over all EXIF data fields
    for tag_id in exifdata:
        # get the tag name, instead of human unreadable tag id
        tag = TAGS.get(tag_id, tag_id)
        data = exifdata.get(tag_id)
        # decode bytes 
        if isinstance(data, bytes):
            data = data.decode()
        #print(f"{tag:25}: {data}")
        if(tag=='BrightnessValue'):
          img_mdata[0]=data[0]/data[1]
        
        if(tag=='ShutterSpeedValue'):
          img_mdata[1]=data[0]/data[1]

        if(tag=='ISOSpeedRatings'):
          img_mdata[2]=data

    df.loc[i] = [file[len_str_remove:]] + [img_mdata[0]] + [img_mdata[1]] + [img_mdata[2]]
    i+=1
    if(i==150):break

print(df.shape)
df
```
