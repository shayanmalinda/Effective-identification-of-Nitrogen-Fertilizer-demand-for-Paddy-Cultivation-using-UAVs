```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

image = cv.imread("IMG_20210721_105517 (2).jpg")
rgbImage = cv.cvtColor(image, cv.COLOR_BGR2RGB)
rgbImageCopy = cv.cvtColor(image, cv.COLOR_BGR2RGB)

hsvImage = cv.cvtColor(image, cv.COLOR_BGR2HSV)

low_green = np.array([25, 50, 70])
high_green = np.array([89, 255, 255])
mask = cv.inRange(hsvImage, low_green, high_green)


plt.imshow(rgbImage)
plt.show()

plt.imshow(mask,'gray')
plt.show()

x = len(rgbImageCopy)
y = len(rgbImageCopy[0])

for i in range(x):
  for j in range(y):
      if mask[i][j] == 0:
        rgbImageCopy[i][j][0] = 255
        rgbImageCopy[i][j][1] = 255
        rgbImageCopy[i][j][2] = 255
      else:
        rgbImageCopy[i][j][0] = rgbImageCopy[i][j][0]
        rgbImageCopy[i][j][1] = rgbImageCopy[i][j][1]
        rgbImageCopy[i][j][2] = rgbImageCopy[i][j][2]

plt.imshow(rgbImage)
plt.show()

plt.imshow(rgbImageCopy)
plt.show()
```
