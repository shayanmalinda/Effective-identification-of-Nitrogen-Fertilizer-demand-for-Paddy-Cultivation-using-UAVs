```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

image = cv.imread("IMG_20210721_103634.jpg")
rgbImage = cv.cvtColor(image, cv.COLOR_BGR2RGB)
rgbImageCopy = cv.cvtColor(image, cv.COLOR_BGR2RGB)

hsvImage = cv.cvtColor(image, cv.COLOR_BGR2HSV)

low_green = np.array([30, 50, 76])
high_green = np.array([80, 255, 255])
mask = cv.inRange(hsvImage, low_green, high_green)
maskInvert = np.invert(mask)

plt.imshow(rgbImage)
plt.show()

plt.imshow(mask,'gray')
plt.show()

b,g,r = cv.split(rgbImage)

b = np.bitwise_or(maskInvert, b)
g = np.bitwise_or(maskInvert, g)
r = np.bitwise_or(maskInvert, r)

result = cv.merge([b,g,r])

plt.imshow(rgbImage)
plt.show()

plt.imshow(result)
plt.show()
```
