```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

image = cv.imread("IMG_20210721_095600.jpg")
rgbImage = cv.cvtColor(image, cv2.COLOR_BGR2RGB)
rgbImageCopy = cv.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
hsvImage = cv.cvtColor(image, cv.COLOR_BGR2HSV)

low_green = np.array([30, 0, 0])
high_green = np.array([90, 255, 255])
mask = cv.inRange(hsvImage, low_green, high_green)


plt.imshow(rgbImage)
plt.show()

plt.imshow(mask,'gray')
plt.show()
```

```python
x = len(rgbImageCopy)
y = len(rgbImageCopy[0])

for i in range(x):
  for j in range(y):
      if mask[i][j] == 0:
        rgbImageCopy[i][j][0] = 255
        rgbImageCopy[i][j][1] = 255
        rgbImageCopy[i][j][2] = 255
      else:
        rgbImageCopy[i][j][0] = rgbImageCopy[i][j][0]
        rgbImageCopy[i][j][1] = rgbImageCopy[i][j][1]
        rgbImageCopy[i][j][2] = rgbImageCopy[i][j][2]

plt.imshow(rgbImage)
plt.show()

plt.imshow(rgbImageCopy)
plt.show()
```
